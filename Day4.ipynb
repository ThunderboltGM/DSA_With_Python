<h2 style="color:yellow;text-align:center">BINARY TREE</h2>

```python
# binary tree
from collections import deque
class TreeNode:
    def __init__(self,val):
        self.val = int(val)
        self.left = None
        self.right = None

def buildTree(value):
    if not value or value[0] == 'null':
        return None
    root = TreeNode(value[0])
    queue = deque([root])
    i = 1
    
    while queue and i<len(value):
        curr = queue.popleft()

        if i<len(value) and value[i] != 'null':
            curr.left = TreeNode(value[i])
            queue.append(curr.left)
        i += 1

        if i<len(value) and value[i] != 'null':
            curr.right = TreeNode(value[i])
            queue.append(curr.right)
        i += 1
    return root

def findLeft(root):
    if not root:
        return None
    queue = deque([root])
    leftmost = root.val

    while queue:
        levelSize = len(queue)
        for i in range(levelSize):
            node = queue.popleft()
            if i == 0:
                leftmost = node.val
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
    return leftmost

size = int(input())
value = list(map(int,input().split()))
root = buildTree(value)
print(findLeft(root))
```

```python
from collections import deque
class TreeNode:
    def __init__(self,val):
        self.val = int(val)
        self.left = None
        self.right = None

def buildTree(value):
    if not value or value[0] == 'null':
        return None
    root = TreeNode(value[0])
    queue = deque([root])
    i = 1
    
    while queue and i<len(value):
        curr = queue.popleft()
        if i<len(value) and value[i] !='null':
            curr.left = TreeNode(value[i])
            queue.append(curr.left)
        i+=1
        if i<len(value) and value[i] != 'null':
            curr.right = TreeNode(value[i])
            queue.append(curr.right)
        i+=1
        return root
    
def findCommon(root,p,q):
    if root is None or root.val == p or root.val == q:
        return root
    left = findCommon(root.left,p,q)
    right = findCommon(root.right,p,q)
    if left and right:
        return root
    return left if left else right
    
size = int(input())
values = input().split()
p,q = map(int,input().split())
root = buildTree(values)
res = findCommon(root,p,q)
print(res.val)
```

```python
# house robbing problem
size = int(input())
arr = list(map(int, input().split()))

def maxMoney(houses):
    n = len(houses)
    if n == 0:
        return 0
    if n == 1:
        return houses[0]
    
    dp = [0] * n
    dp[0] = houses[0]
    dp[1] = max(houses[0], houses[1])
    
    for i in range(2, n):
        dp[i] = max(houses[i] + dp[i-2], dp[i-1])
    
    return dp[-1]

print(maxMoney(arr))
```

```python
from collections import deque

class TreeNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
        
def buildTree(values):
    if not values or values[0] == 'null':
        return None
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    
    while queue and i<len(values):
        
        curr  = queue.popleft()
        
        if i<len(values) and values[i] != 'null':
            curr.left = TreeNode(values[i])
            queue.append(curr.left)
        i+=1
        
        if i<len(values) and values[i] != 'null':
            curr.right = TreeNode(values[i])
            queue.append(curr.right)
        i+=1
    
    return root

def pathCalculate(root,path,paths):
    if not root:
        return None
    
    path.append(str(root.data))
    
    if root.left is None and root.right is None:
        paths.append(str("".join(path)))
    else:
        pathCalculate(root.left,path,paths)
        pathCalculate(root.right,path,paths)
        
    path.pop()
    
    return sum(int(path) for path in paths)
    
    
size = int(input())
values = input().split()
root = buildTree(values)
paths = []
print(pathCalculate(root,[],paths))
```

<h2 style="color:yellow;text-align:center">GRAPH</h2>

```python
# to check if the graph is connected graph or not
v,e = map(int,input().split())
edges = [tuple(map(int,input().split())) for _ in range(e)]
def isConnected(v,e,edges):
    adj = [[] for _ in range(v)]
    for a,b in edges:
        adj[a].append(b)
        adj[b].append(a)
    
    vis = [False]*v
    
    def dfs(v):
        vis[v] = True
        for neighbour in adj[v]:
            if not vis[neighbour]:
                dfs(neighbour)
    
    if v>0:
        dfs(0)
        
    return all(vis)

print('true' if isConnected(v,e,edges) else 'false')
```
