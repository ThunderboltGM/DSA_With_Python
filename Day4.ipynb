```python
# binary tree
from collections import deque
class TreeNode:
    def __init__(self,val):
        self.val = int(val)
        self.left = None
        self.right = None

def buildTree(value):
    if not value or value[0] == 'null':
        return None
    root = TreeNode(value[0])
    queue = deque([root])
    i = 1
    
    while queue and i<len(value):
        curr = queue.popleft()

        if i<len(value) and value[i] != 'null':
            curr.left = TreeNode(value[i])
            queue.append(curr.left)
        i += 1

        if i<len(value) and value[i] != 'null':
            curr.right = TreeNode(value[i])
            queue.append(curr.right)
        i += 1
    return root

def findLeft(root):
    if not root:
        return None
    queue = deque([root])
    leftmost = root.val

    while queue:
        levelSize = len(queue)
        for i in range(levelSize):
            node = queue.popleft()
            if i == 0:
                leftmost = node.val
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
    return leftmost

size = int(input())
value = list(map(int,input().split()))
root = buildTree(value)
print(findLeft(root))
```

```python
from collections import deque
class TreeNode:
    def __init__(self,val):
        self.val = int(val)
        self.left = None
        self.right = None

def buildTree(value):
    if not value or value[0] == 'null':
        return None
    root = TreeNode(value[0])
    queue = deque([root])
    i = 1
    
    while queue and i<len(value):
        curr = queue.popleft()
        if i<len(value) and value[i] !='null':
            curr.left = TreeNode(value[i])
            queue.append(curr.left)
        i+=1
        if i<len(value) and value[i] != 'null':
            curr.right = TreeNode(value[i])
            queue.append(curr.right)
        i+=1
        return root
    
def findCommon(root,p,q):
    if root is None or root.val == p or root.val == q:
        return root
    left = findCommon(root.left,p,q)
    right = findCommon(root.right,p,q)
    if left and right:
        return root
    return left if left else right
    
size = int(input())
values = input().split()
p,q = map(int,input().split())
root = buildTree(values)
res = findCommon(root,p,q)
print(res.val)
```
